```python
import numpy as np
np.set_printoptions(threshold=10000,suppress=True)
import pandas as pd
import warnings
import matplotlib.pyplot as plt
warnings.filterwarnings('ignore')
```

```python
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler

# Charger les données (remplacez 'data.csv' par votre fichier)
data = pd.read_csv("villes.csv", sep=';')
X = data.iloc[:, 1:13].values
labels = data.iloc[:, 0].values

# Normalisation des données
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Application de l'ACP pour réduction de dimension à 2 composantes
pca = PCA(2)
X_pca = pca.fit_transform(X_scaled)

# Application de KMeans avec 3 clusters
kmeans = KMeans(n_clusters=3, random_state=42, n_init=10)
clusters = kmeans.fit_predict(X_pca)

# Affichage des résultats
plt.figure(figsize=(10, 6))
for cluster in np.unique(clusters):
    plt.scatter(X_pca[clusters == cluster, 0], X_pca[clusters == cluster, 1], label=f'Cluster {cluster}')

# Ajouter les labels des villes
for i, label in enumerate(data.index):
    plt.text(X_pca[i, 0], X_pca[i, 1], label, fontsize=9, ha='right')

plt.xlabel("Composante principale 1")
plt.ylabel("Composante principale 2")
plt.title("Clustering des villes avec KMeans")
plt.legend()
plt.show()
```
