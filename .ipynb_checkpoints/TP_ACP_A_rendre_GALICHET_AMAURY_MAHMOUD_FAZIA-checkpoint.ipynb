```python
data = pd.read_csv('./villes.csv', sep=';')
X = data.iloc[:, 1:13].values
labels = data.iloc[:, 0].values
data
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
# Standardize the features
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Apply PCA
pca = PCA()
X_pca = pca.fit_transform(X_scaled)
coefficients = pca.components_
explained_variance_ratio = np.cumsum(pca.explained_variance_ratio_)
# Détermination du nombre de composantes nécessaires
num_components = np.argmax(explained_variance_ratio >= 0.90) + 1
print(f"\n Nombre de composantes principales nécessaires pour expliquer 90% de la variance : {num_components}\n")
```

En exemple on retrouve Nice qui a un PC1 élevé et un PC2 faible (été chaud et hiver doux + faibles variation saisonnières).

####  Villes océaniques et tempérées (PC1 moyen, PC2 élevé)
En exemple on retrouve Nantes, qui a un PC1 moyen et un PC2 élevé (été doux et hiver doux + faible amplitude thermique).
Les villes océaniques ont une température relativement stable, elles ne seront ni globalement chaudes ni globalement froides.

####  Villes froides et continentales (PC1 faible, PC2 faible)
En exemple Besançon , qui a un PC1 faible et un PC2 faible (été modéré, hiver froid + forte amplitude thermique et forte variation saisonnière).

---

#### Interprétation des axes :
PC1 : Différencie les villes chaudes (Sud) des villes froides (Nord/Est).
PC2 : Mesure l’amplitude thermique, séparant climats stables (océanique) et climats contrastés (continental).

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA

def analyse_acp(file_path):
    """
    Effectue une Analyse en Composantes Principales (ACP) sur un fichier CSV contenant des données numériques.

    Paramètre :
        - file_path (str) : Chemin du fichier CSV.

    Résultats :
        - Affichage du nombre de composantes principales nécessaires pour expliquer ≥ 90% de la variance.
        - Graphique de la variance expliquée.
        - Projection des observations dans le plan des deux premières composantes.
    """

    # Chargement des données
    data = pd.read_csv(file_path, sep=';')

    # Séparation des labels (1ère colonne) et des valeurs numériques
    labels = data.iloc[:, 0].values
    X = data.iloc[:, 1:].values  # On prend toutes les colonnes sauf la première

    # Standardisation des données (centrées-réduites)
    scaler = StandardScaler()
    X_scaled = scaler.fit_transform(X)

    #  Application de l'ACP
    pca = PCA()
    X_pca = pca.fit_transform(X_scaled)
    explained_variance_ratio = np.cumsum(pca.explained_variance_ratio_)
    coefficients = pca.components_

    # Détermination du nombre de composantes nécessaires
    num_components = np.argmax(explained_variance_ratio >= 0.90) + 1
    print(f"\n Nombre de composantes principales nécessaires pour expliquer 90% de la variance : {num_components}\n")

    # Graphique de la variance expliquée
    plt.figure(figsize=(10, 6))
    plt.plot(range(1, len(explained_variance_ratio) + 1), explained_variance_ratio, 'bo-')
    plt.xlabel('Nombre de Composantes')
    plt.ylabel('Variance Cumulative Expliquée')
    plt.title('Variance Expliquée vs Nombre de Composantes')
    plt.grid(True)
    plt.show()

    # Projection des données dans le plan des deux premières composantes
    plt.figure(figsize=(12, 8))
    plt.scatter(X_pca[:, 0], X_pca[:, 1], color='blue')

    # Ajout des labels (si disponibles)
    for i, label in enumerate(labels):
        plt.annotate(label, (X_pca[i, 0], X_pca[i, 1]), xytext=(-0.2, 0.2), textcoords='offset points')

    plt.xlabel('Première Composante Principale')
    plt.ylabel('Deuxième Composante Principale')
    plt.title('Projection des Observations dans le Plan ACP')
    plt.grid(True)
    plt.show()

    # Affichage du ratio de variance expliquée
    print("\n Ratio de variance expliquée par composante :")
    for i, ratio in enumerate(pca.explained_variance_ratio_):
        print(f"PC{i+1}: {ratio:.3f}")

    # Affichage des coefficients par axes du graphe
    print("\nCoefficients pour PC1 :")
    print(coefficients[0])
    
    # Affichage des coefficients par axes du graphe
    print("\nCoefficients pour PC2 :")
    print(coefficients[1])
    

```

### Analyse des résultats:

La première chose que l'on remarque est le fait qu'il faut 4 axes pour conserver un minimum de 90% de l'information. Antrement dit, le graphe présenté ci-dessus n'est pas un modèle adapté aux données du fichier (projection sur deux axes).

PC1 : représente le niveau général de criminalité.

#### Typologie des états (selon leur position sur le graphique) :
Les états à faible criminalité ont un PC1 négatif ( se situent à gauche dans le plan).

Les états à criminalité modérée ont un PC1 proche de 0 ( se situent au centre). 

Les états à forte criminalité ont un PC1 positif ( se situent à droite dans le plan).

#### Interprétation :

On constate une perte d'information conséquente par rapport aux données sur les températures des villes qui nécessitait une projection sur deux axes pour conserver 90% de l'information.

Nous pouvons expliquer cela par les multiples types de variables qui ne sont pas comparables entre elles. En effet, les températures en fonction des mois sont du même type (température), donc comparables. Dans le cas des délits et crimes par états, ils ne sont pas comparables entre eux (les types de crimes sont nombreux pour être interprétés rapidement).

Il n'est pas possible d'identifier visuellement des groupes d'êtats avec une certaine caractéristique.
